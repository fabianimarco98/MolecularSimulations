```python
with open("rmsf.xvg","r") as f:
    l=f.readlines()
diz={}
for el in l:
    if not (el.startswith("#") or el.startswith("@")):
        s=el.strip()
        temp=s.split()
        diz[temp[0]]=float(temp[1])

sorted_by_value = sorted(diz.items(), key=lambda item: item[1], reverse=True)
sorted_diz=dict(sorted_by_value)
sorted_diz
```
